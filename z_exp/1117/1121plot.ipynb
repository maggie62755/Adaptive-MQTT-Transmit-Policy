```python
data_range=[]
def object_to_int(input_data):
    input_data[:,0] = input_data[:,0].astype(int)
    input_data[:,1] = input_data[:,1].astype(int)
    input_data=input_data[input_data[:,0].argsort()]
    data_range.append((max(input_data[:,1])-min(input_data[:,1])));

data1 = np.loadtxt('/home/maggie/mosq2.0.12/1108mosq/z_exp/1117/bro_to_sub1.out.prev', dtype=object)
data2 = np.loadtxt('/home/maggie/mosq2.0.12/1108mosq/z_exp/1117/bro_to_sub2.out.prev', dtype=object)
data3 = np.loadtxt('/home/maggie/mosq2.0.12/1108mosq/z_exp/1117/bro_to_sub3.out.prev', dtype=object)
data4 = np.loadtxt('/home/maggie/mosq2.0.12/1108mosq/z_exp/1117/bro_to_sub4.out.prev', dtype=object)
object_to_int(data1)
object_to_int(data2)
object_to_int(data3)
object_to_int(data4)
```

```python
range=max(data_range)
print(data_range)
print(range)


# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(8, 4),dpi=150)

# make a plot
ax.plot(data1[:,0], data1[:,1] , alpha=0.7,color="blue")
ax.plot(data2[:,0], data2[:,1] , alpha=0.7,color="cyan")
ax.set_xlabel("send time")
ax.set_ylabel("latency (us)",color="blue")
# ax.set_ylim(bottom=min(data1[:,1])-10000,top=min(data1[:,1])+range)


ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.scatter(data3[:,0], data3[:,1], marker = 'o', alpha=0.7,color="red")
# ax2.scatter(data4[:,0], data4[:,1], marker = 'o', alpha=0.7,color="pink")
ax2.set_ylabel("latency (s)",color="red")
# ax2.set_ylim(bottom=min(data3[:,1])-10000,top=min(data3[:,1])+range)

plt.xlim(int(data1[0][0]),int(data1[data1[:,0].size-1][0]))
plt.ylabel("latency (us)") # y label
plt.show()
```
